## Separando os dados da Paraíba e alunos do curso de engenharia de computação I

Vamos separar apenas os dados da Paraíba e alunos do curso de engenharia de computação I, para teremos:

- `CO_UF_CURSO` = 25
- `CO_GRUPO` = 6411

Separando os códigos do grupo de Engenharia de Computação I:

```python
co_grupo_eng_comp = microdados2023_arq1[microdados2023_arq1["CO_GRUPO"] == 6411]
co_grupo_eng_comp
```

Separando os alunos da Paraíba (PB):

```python
co_grupo_eng_comp_pb = co_grupo_eng_comp[co_grupo_eng_comp["CO_UF_CURSO"] == 25].reset_index(drop=True)
co_grupo_eng_comp_pb
```

```python
co_cursos_eng_comp_pb = co_grupo_eng_comp_pb["CO_CURSO"].unique()
co_cursos_eng_comp_pb
```

Como os diferentes datasets utilizados na análise não possuem um identificador único comum entre si, foi necessário criar uma chave artificial para permitir a junção entre eles. Para isso, os dados foram inicialmente ordenados pela coluna `CO_CURSO`, garantindo uma correspondência consistente entre as observações. Em seguida, foi criada uma nova coluna chamada `ID`, combinando o código do curso (`CO_CURSO`) com o índice da linha já resetado. Esse identificador composto assegura a unicidade dentro do escopo da análise e permite a fusão dos datasets de forma alinhada.

```python
co_grupo_eng_comp_pb = co_grupo_eng_comp_pb.sort_values(by="CO_CURSO")
co_grupo_eng_comp_pb = co_grupo_eng_comp_pb.reset_index(drop=True)
co_grupo_eng_comp_pb["ID"] = co_grupo_eng_comp_pb["CO_CURSO"].astype(str) + "_" + co_grupo_eng_comp_pb.index.astype(str)
co_grupo_eng_comp_pb
```

Filtrando todos os datasets de perguntas:

```python
dfs_filtrados_eng_comp_pb = []

for df in dfs:
    df_filtrado = df[df["CO_CURSO"].isin(co_cursos_eng_comp_pb)]
    df_filtrado = df_filtrado.sort_values(by="CO_CURSO").reset_index(drop=True)
    df_filtrado["ID"] = df_filtrado["CO_CURSO"].astype(str) + "_" + df_filtrado.index.astype(str)
    df_filtrado = df_filtrado.drop(columns=["NU_ANO", "CO_CURSO"])
    cols = df_filtrado.columns
    df_filtrado = df_filtrado[[cols[1], cols[0]]]
    dfs_filtrados_eng_comp_pb.append(df_filtrado)

dfs_filtrados_eng_comp_pb
```

## Concatenando os datasets:

Unindo o questionário:

```python
df_merged_questionario = dfs_filtrados_eng_comp_pb[0]
for df in dfs_filtrados_eng_comp_pb[1:]:
    df_merged_questionario = df_merged_questionario.merge(df, on="ID", how="outer")
```

```python
df_merged_questionario
```

Unindo o questionário com os demais datasets:

```python
dfs_merge = [co_grupo_eng_comp_pb, notas_eng_comp_pb, sexo_eng_comp_pb, idade_eng_comp_pb, df_merged_questionario]

# Inicializa com o primeiro dataframe da lista
eng_comp_pb_enade23 = dfs_merge[0]

# Itera do segundo em diante, para evitar merge do df com ele mesmo
for df in dfs_merge[1:]:
    eng_comp_pb_enade23 = eng_comp_pb_enade23.merge(df, on="ID", how="outer")

eng_comp_pb_enade23
```

Preenchendo os valores faltantes com a opção mais frequente (moda):

```python
colunas_questoes = [f'QE_I{str(i).zfill(2)}' for i in range(1,27)]

for col in colunas_questoes:
    moda = df[col].mode()[0] # valor mais frequente na coluna
    eng_comp_pb_enade23[col].fillna(moda, inplace=True)
```

Preenchendo os valores que faltam na coluna das notas:

```python
# Verificar quantos valores faltam
print(eng_comp_pb_enade23['NT_GER'].isna().sum())

# preenchendo com a mediana
mediana = eng_comp_pb_enade23['NT_GER'].median()
eng_comp_pb_enade23['NT_GER'].fillna(mediana, inplace=True)
```

Os dados faltantes nas perguntas categóricas foram preenchidos pela moda, que representa o valor mais comum nas respostas. Para a nota geral (NT_GER), utilizei a mediana para preencher os valores ausentes, pois ela é menos sensível a valores extremos e melhor representa a tendência central dos dados numéricos.

## Sumário dos dados do dataset criado:

- `NU_ANO`: Ano de realização do exame.
- `CO_CURSO`: Código do curso no Enade.
- `CO_IES`: Código da IES (e-MEC).
- `CO_CATEGAD`: Código da categoria administrativa da IES.
- `CO_ORGACAD`: Código da organização acadêmica da IES.
- `CO_GRUPO`: Código da Área de enquadramento do curso no Enade.
- `CO_MODALIDADE`: Código da Modalidade de Ensino.
- `CO_MUNIC_CURSO`: Código do município de funcionamento do curso.
- `CO_UF_CURSO`: Código da UF de funcionamento do curso.
- `CO_REGIAO_CURSO`: Código da região de funcionamento do curso.
- `ID`: Identificador dos estudantes criado artificialmente.
- `NT_GER`: Nota bruta da prova - Média ponderada da formação geral (25%) e componente específico (75%). (valor de 0 a 100).
- `TP_SEXO`: Sexo.
- `NU_IDADE`: Idade do inscrito em 26/11/2023.
- `QE_I01` - `QE_I26`: Questionário socioeconômico do estudante.
